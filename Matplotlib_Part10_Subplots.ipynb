So far we have been importing pyplot object and working with it. 

Naming Convention :

*    **Figure** - is an container holding our plots
*    **Axes** - These are the actual plots

When working with multiple plots we should always work in object oriented way.

Lets plot the above data using subplots. 

```
import pandas as pd
from matplotlib import pyplot as plt


plt.style.use('seaborn')

!wget --no-check-certificate \
'https://raw.githubusercontent.com/bgswaroop/python-matplotlib/master/data/part5_data.csv' \
-O '/tmp/part5_data.csv'

data = pd.read_csv('/tmp/part5_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, ax = plt.subplots() # by default we have only one row and one column
# now ax represents our plot (single plot)

ax.plot(ages, py_salaries, label='Python')
ax.plot(ages, js_salaries, label='JavaScript')
ax.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
ax.legend()

# Note that the following have now changed to set_
ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

Lets now plot the charts separately

```
import pandas as pd
from matplotlib import pyplot as plt


plt.style.use('seaborn')

!wget --no-check-certificate \
'https://raw.githubusercontent.com/bgswaroop/python-matplotlib/master/data/part5_data.csv' \
-O '/tmp/part5_data.csv'

data = pd.read_csv('/tmp/part5_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

# Here we are unpacking the axes into ax1 and ax2
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1) # by default we have only one row and one column

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

Since we are having only one kind of data so lets have a common title and also having x axis label only on the bottom plot is sufficient. So lets make these changes.

```
import pandas as pd
from matplotlib import pyplot as plt


plt.style.use('seaborn')

!wget --no-check-certificate \
'https://raw.githubusercontent.com/bgswaroop/python-matplotlib/master/data/part5_data.csv' \
-O '/tmp/part5_data.csv'

data = pd.read_csv('/tmp/part5_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

# Here we are unpacking the axes into ax1 and ax2
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1) # by default we have only one row and one column

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

Now we can also have only common x tick marks **sharex**

```
import pandas as pd
from matplotlib import pyplot as plt


plt.style.use('seaborn')

!wget --no-check-certificate \
'https://raw.githubusercontent.com/bgswaroop/python-matplotlib/master/data/part5_data.csv' \
-O '/tmp/part5_data.csv'

data = pd.read_csv('/tmp/part5_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

# Here we are unpacking the axes into ax1 and ax2
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True) 

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

We can also have multiple figures instead of multiple plots within each figure. In PyCharm this will open two separate windows for each chart.

We can also save these figures to disk

```
import pandas as pd
from matplotlib import pyplot as plt


plt.style.use('seaborn')

!wget --no-check-certificate \
'https://raw.githubusercontent.com/bgswaroop/python-matplotlib/master/data/part5_data.csv' \
-O '/tmp/part5_data.csv'

data = pd.read_csv('/tmp/part5_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

# Here we are unpacking the axes into ax1 and ax2
fig1, ax1 = plt.subplots() 
fig2, ax2 = plt.subplots() 

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()

# Saving figures to current directory
fig1.savefig('fig1.png')
fig1.savefig('fig2.png')
```
