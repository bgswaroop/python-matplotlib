Lets write some quick starter code for pie a plot


```
from matplotlib import pyplot as plt

# Apply your plt style here
plt.style.use('fivethirtyeight')

# Create pie plot
slices = [60, 40]
plt.pie(slices)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

Isn't that simple ! Now, lets add labels to this plot.

Note that the numbers in the slices need not add upto 100.

```
from matplotlib import pyplot as plt

# Apply your plt style here
plt.style.use('fivethirtyeight')

# Create pie plot
slices = [120, 80]
labels = ['Sixty', 'Forty']

plt.pie(slices, labels=labels)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

Well we can format our chart a little bit


*   Change the border color for this pie chart - **wedgeprops -> edgecolor**
*   Manually set the colors for each pie. Set this as a list.



```
from matplotlib import pyplot as plt

# Apply your plt style here
plt.style.use('fivethirtyeight')

# Create pie plot
slices = [120, 80, 30, 20]
labels = ['Sixty', 'Forty', "Extra1", "Extra2"]
colors = ['#008fd5', '#fc4f30', '#e5ae37', "#6d904f"]

plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'}, colors=colors)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

Let's try plotting pie charts on real world data. 2019 StackOverflow developers survey.

```
from matplotlib import pyplot as plt

# Apply your plt style here
plt.style.use('fivethirtyeight')

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
#plt.tight_layout()
plt.show()
```

Pie chart does look too crowded, and so lets plot only the top five languages.



*   There is another interesting feature in pie charts called - **explode**. We can use this to emphasize some particular pie's.
*   We could also add **shadow** to our chart 

```
from matplotlib import pyplot as plt

# Apply your plt style here
plt.style.use('fivethirtyeight')

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

explode = [0, 0, 0, 0.1, 0] 
# 0.1 indicates that we want to shift out the pie by 0.1 times the radius of pie.


plt.pie(slices, labels=labels, explode = explode, shadow=True,
        wedgeprops={'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
#plt.tight_layout()
plt.show()
```

Let us now look at some additional common tricks with pie charts.


*   We can also set the **starting angle**. This will rotate the pie chart.
*   We can automatically **calculate the percentages** and display them on the chart.







```
from matplotlib import pyplot as plt

# Apply your plt style here
plt.style.use('fivethirtyeight')

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

explode = [0, 0, 0, 0.1, 0] 
# 0.1 indicates that we want to shift out the pie by 0.1 times the radius of pie.


plt.pie(slices, labels=labels, explode = explode, shadow=True, startangle=90,
        autopct='%1.1f%%', wedgeprops={'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
#plt.tight_layout()
plt.show()
```
