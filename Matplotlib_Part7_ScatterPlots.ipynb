These are useful to visualize some outliers.

Lets format the scatter plots :
*    **size** - change the size of the marker
*    **color** 
*    **marker** style
*    **edgecolor** - edge of the marker
*    **linewidth** - of the edge
*    **alpha** - transparency for the color


```
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y, s=100, c='green', edgecolor='black', linewidth=1, alpha=0.75)

plt.tight_layout()
plt.show()
```

Lets now load some real data from youtube statistics. The csv_file contains the details for each youtube video : view_counts, likes and ratio.

This data is collected for top 200 videos on the trending page.

```
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

!wget --no-check-certificate \
'https://raw.githubusercontent.com/bgswaroop/python-matplotlib/master/data/part7_data.csv' \
-O '/tmp/part7_data.csv'

data = pd.read_csv('/tmp/part7_data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)

plt.title('Trending videos on Youtube')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```

The plot looks cluttered to the bottom left, because we have one outlier on the top right. So lets use log scale for the plot.

```
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

!wget --no-check-certificate \
'https://raw.githubusercontent.com/bgswaroop/python-matplotlib/master/data/part7_data.csv' \
-O '/tmp/part7_data.csv'

data = pd.read_csv('/tmp/part7_data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)
plt.xscale('log')
plt.yscale('log')

plt.title('Trending videos on Youtube')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```

We can also use 2d scatter plots, to include a third dimension of information using color.

```
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

!wget --no-check-certificate \
'https://raw.githubusercontent.com/bgswaroop/python-matplotlib/master/data/part7_data.csv' \
-O '/tmp/part7_data.csv'

data = pd.read_csv('/tmp/part7_data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75, 
            c=ratio, cmap='summer')
cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending videos on Youtube')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```
