The data file contains developers age and their median salary for py, js and all_devs.


```
import pandas as pd
from matplotlib import pyplot as plt

!wget --no-check-certificate \
'https://raw.githubusercontent.com/bgswaroop/python-matplotlib/master/data/part5_data.csv' \
-O '/tmp/part5_data.csv'

data = pd.read_csv('/tmp/part5_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

# overall_median = 57287

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

Now, Lets fill the area under the python plot, using **plt.fill_between**

```
import pandas as pd
from matplotlib import pyplot as plt

!wget --no-check-certificate \
'https://raw.githubusercontent.com/bgswaroop/python-matplotlib/master/data/part5_data.csv' \
-O '/tmp/part5_data.csv'

data = pd.read_csv('/tmp/part5_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

This does not look appealing, lets add some transparancy to this fill.

```
import pandas as pd
from matplotlib import pyplot as plt

!wget --no-check-certificate \
'https://raw.githubusercontent.com/bgswaroop/python-matplotlib/master/data/part5_data.csv' \
-O '/tmp/part5_data.csv'

data = pd.read_csv('/tmp/part5_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries, alpha=0.25)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

Right now we are filling between zero and the py_salaries.

But, we could use some additional conditions !

```
import pandas as pd
from matplotlib import pyplot as plt

!wget --no-check-certificate \
'https://raw.githubusercontent.com/bgswaroop/python-matplotlib/master/data/part5_data.csv' \
-O '/tmp/part5_data.csv'

data = pd.read_csv('/tmp/part5_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median, alpha=0.25)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

In the above plot, we now know where the median is and how far the python salaries are away from this median.

*    where - where clause to conditionally fill an area
*    interpolate - To avoid clipping


```
import pandas as pd
from matplotlib import pyplot as plt

!wget --no-check-certificate \
'https://raw.githubusercontent.com/bgswaroop/python-matplotlib/master/data/part5_data.csv' \
-O '/tmp/part5_data.csv'

data = pd.read_csv('/tmp/part5_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median, 
                 where=(py_salaries > overall_median), 
                 interpolate=True, alpha=0.25)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

we can have multiple fill betweens to highlight different regions in our plot.

```
import pandas as pd
from matplotlib import pyplot as plt

!wget --no-check-certificate \
'https://raw.githubusercontent.com/bgswaroop/python-matplotlib/master/data/part5_data.csv' \
-O '/tmp/part5_data.csv'

data = pd.read_csv('/tmp/part5_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median, 
                 where=(py_salaries > overall_median), 
                 interpolate=True, alpha=0.25)

plt.fill_between(ages, py_salaries, overall_median, 
                 where=(py_salaries <= overall_median), 
                 interpolate=True, alpha=0.25)
plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

We could also fill between the area between two line plots

```
import pandas as pd
from matplotlib import pyplot as plt

!wget --no-check-certificate \
'https://raw.githubusercontent.com/bgswaroop/python-matplotlib/master/data/part5_data.csv' \
-O '/tmp/part5_data.csv'

data = pd.read_csv('/tmp/part5_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, dev_salaries, 
                 where=(py_salaries > dev_salaries), 
                 interpolate=True, alpha=0.25)

plt.fill_between(ages, py_salaries, dev_salaries, 
                 where=(py_salaries <= dev_salaries), 
                 interpolate=True, alpha=0.25)
plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

We can also have labels for fill between

```
import pandas as pd
from matplotlib import pyplot as plt

!wget --no-check-certificate \
'https://raw.githubusercontent.com/bgswaroop/python-matplotlib/master/data/part5_data.csv' \
-O '/tmp/part5_data.csv'

data = pd.read_csv('/tmp/part5_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, dev_salaries, 
                 where=(py_salaries > dev_salaries), 
                 interpolate=True, alpha=0.25, label='above avg')

plt.fill_between(ages, py_salaries, dev_salaries, 
                 where=(py_salaries <= dev_salaries), 
                 interpolate=True, alpha=0.25, label='below avg')
plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
